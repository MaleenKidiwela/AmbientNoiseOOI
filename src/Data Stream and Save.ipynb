```python
import os
from obspy.clients.fdsn import Client
from obspy import UTCDateTime

def fetch_and_save_waveforms(year, network, station):
    client = Client("IRIS")
    location = ""
    channels = ["HHZ", "HHN", "HHE"]
    
    start_date = UTCDateTime(f"{year}-01-01")
    end_date = UTCDateTime(f"{year}-12-31")

    # Calculate total days including the last day
    total_days = int((end_date - start_date) / 86400) + 1  # Use regular division and convert to int

    for current_day in range(total_days):
        day = start_date + current_day * 86400
        doy = day.julday  # Day of the year

        for channel in channels:
            try:
                waveform = client.get_waveforms(network, station, location, channel, day, day + 86400)
                day_dir = f"/data/wsd02/maleen_data/NEPTUNE-Data/{station}/{year}/{doy:03d}"
                os.makedirs(day_dir, exist_ok=True)
                filename = os.path.join(day_dir, f"{station}.{network}.{year}.{doy:03d}.{channel}")
                waveform.write(filename, format="MSEED")
                print(f"Data written for {filename}")
            except Exception as e:
                print(f"Failed to fetch or write data for {channel} on {day}: {e}")

# Example usage
fetch_and_save_waveforms(2024, "NV", "NC89")
```
